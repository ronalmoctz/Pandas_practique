```python
import pandas as pd
```

```python
df_players = pd.read_csv('dataset/players_20.csv')
```

```python
df_players
```

```python
df_players = df_players.set_index('short_name', inplace=True)
```

```python
# Select Column
df_players[['short_name','long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

# **1.- Select with values methos 'loc' -> [row_name, column_name]**

```python
df_players.loc['L. Messi']
```

```python
# Get size of L. Messi
df_players.loc['L. Messi', 'height_cm']
```

```python
# Get the weight of Cristiano Ronaldo
df_players.loc['Cristiano Ronaldo', 'weight_kg']
```

```python
# Get all rows into a specific column
df_players.loc[:, 'height_cm']
# the double dot (:) means all rows
```

```python
# Get all columns into a specific row, example index 'Neymar Jr'
df_players.loc['Neymar Jr',:]
```

## **2.- Select a column with list of values**

```python
# Get data about L. Messi and Cristiano Ronaldo
df_players.loc[['L. Messi', 'Cristiano Ronaldo'], ['age', 'height_cm', 'weight_kg']]
```

## **3.- Select a range of data with one slice arg: start:stop:step (!NOTE: However the slice stop is exclusive**)

```python
# Make slice from names in column
players =['L. Messi', 'Neymar Jr', 'E. Hazard']

df_players.loc[players, 'age':'club']
```

```python
# Do slice to name on the index
columns = ['age', 'dob', 'height_cm', 'weight_kg']

# Get the names of player top1 and top10
df_players.index[:10]

df_players.loc['L. Messi':'M. Salah', columns]
```

```python
df_players.head(10)[columns]
```

## **4.- Select whit conditions**

```python
# one condition: select players with height greater than 180cm
player_height = df_players['height_cm'] > 180

df_players.loc[player_height, 'height_cm'].head(10)
```

```python
# Multiple conditions: select players with height greater than 180cm and their region is Brazil

country_filter = (df_players['height_cm'] >= 180) & (df_players['nationality'] == "Brazil")

df_players.loc[country_filter, ['nationality', 'height_cm']].head(10)
```

# **# Chapter 2 method iLoc,** **Select by position index [row_index, column_index]**

```python
# Get highest of L. Messi
df_players.iloc[0,5]
```

```python
# Get the weight of Cristiano Ronaldo
df_players.iloc[1,6]
```

```python
# Get all rows into the column 'height_cm'
df_players.iloc[:, 5]
```

```python
# Get all the columns that belong to the index 'Neymar Jr'
df_players.iloc[2, :]
```

## **Part three select a range of data with slice args: start:stop:step**

```python
# Do slice to name in names on the index
# players_name = ['L. Messi', 'Neymar Jr']
players_name = df_players[0,2]

# age:club
df_players.iloc[players_name, [1,7]]
```

## **4.- Section with conditions**

```python
# A conditional: select players whit height greater than 180cm
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]
df_players.iloc[df_players['height_cm'] > 180, columns]
```

```python
# select players with height greater than 180cm and their region is Brazil
country_filter = (df_players['height_cm'] >= 180) & (df_players['nationality'] == "Brazil")
df_players.iloc[country_filter, [3,5]]
```
