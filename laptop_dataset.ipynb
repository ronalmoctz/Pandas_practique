```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('dataset/laptop_price.csv')
```

```python
df_laptops
```

```python
import numpy as np
```

```python
# Multiple conditions into a new columns
conditions = [
    df_laptops['Price_euros']> 3000,
    (df_laptops['Price_euros']> 2000) & (df_laptops['Price_euros']<= 3000),
    (df_laptops['Price_euros']> 800) & (df_laptops['Price_euros']<= 2000),
    df_laptops['Price_euros']<= 800
]

values = ['Very Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# Add new column 'level_price'
df_laptops['level_price'] = np.select(conditions, values, default='Unknown')
```

```python

#Show dataframe
df_laptops
```

```python
# Count values in level_price
df_laptops.value_counts('level_price')
```

#### **2.1Exercise:Create Array list based on multiple conditions**

```python
# Array list based on creen size example -> { very big > 16, big > 14, medium > 12, small < 12}
screen_conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <= 14),
    df_laptops['Inches'] <= 12
]

screen_values = ['Very Big', 'Big', 'Medium', 'Small']
```

```python
# Add new column 'screen_size'
df_laptops['screen_size'] = np.select(screen_conditions, screen_values, default='Null')
```

```python
df_laptops
```

```python
#Count the values in screen_size
df_laptops.value_counts('screen_size')
```


## **Example 2: Search the laptops more chep and more expensive each company using the method short_values() and duplicated() (args: keep "first", "last" and False)** ##

```python
# Order dataframe ascending by "Company" and "Price" first cheap and last expensive
sort_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
sort_laptops.value_counts('Company')
```

```python
# Duplicated values in the "Company" column, keep the first (cheap)
dupli_first=sort_laptops.duplicated('Company', keep='first')
```

```python
# show dataframe with values duplicated in the columns "Company"
df_laptops[dupli_first]

# keep 'first'(laptops more cheap of each company
# show dataframe with values not duplicated in the columns "Company"
df_laptops[~dupli_first][['Company', 'Price_euros']]
```

```python
# keep = 'last' (laptops more expensive of each company)
dupli_last = df_laptops.duplicated('Company', keep='last')
```

```python
df_laptops[~dupli_last][['Company', 'Price_euros']]
```

```python
#Keep = False (all laptops more cheap and more expensive of each company)
dupli_all = df_laptops.duplicated('Company', keep=False)
```

```python
df_laptops[~dupli_all][['Company', 'Price_euros']]
```
